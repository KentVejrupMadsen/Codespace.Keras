### Build dataset

```python
counter = 0

if isdir(path_to_build):
    print('deleting directory')
    rmtree(path_to_build)

    print('creating directory')
    mkdir(path_to_build)


def create_if_not_existing(dir):
    path_name = os.path.join(path_to_build, dir)

    if not isdir(path_name):
        os.mkdir(path_name)
        print('creating: ', path_name)

for root, dirs, files in walk(path_to_dataset):
    for file in files:
        counter = counter + 1

        full_path = os.path.join(root, file)

        new_dir_name = full_path[len('/mnt/c/Workspace/symbol-image-dataset/sets/datasets/'):len(root)]
        new_dir_name = new_dir_name.replace('/', '.')

        create_if_not_existing(new_dir_name)

        file_split = file.split('.')
        file_extension = file_split[len(file_split) - 1]

        new_file_name = str(counter) + '.' + file_extension
        path_to_build_file = os.path.join(new_dir_name, new_file_name)
        path_to_build_file = os.path.join(path_to_build, path_to_build_file)

        print(path_to_build_file)

        copy(full_path, path_to_build_file)


```
